# CLASSIC CONTROL: CONTROL THEORY PROBLEM FROM THE CLASSIC RL LITERATURE

```python
import gym # importing gym to create and interact with reinforcement learning environments
import numpy as np
from IPython import display # importing display to show outputs in jupyter notebooks
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def test_env(env: gym.Env) -> None:
    # function to test the environment by taking random actions and showing the output

    env.reset()  # reset the environment to start from the beginning
    done = False  # variable to track if the episode is finished
    img = plt.imshow(env.render(mode='rgb_array'))  # get the first frame of the environment

    while not done:  # keep running until the episode ends
        _, _, done, _ = env.step(env.action_space.sample())  # take a random action in the environment and update the 'done' status

        img.set_data(env.render(mode='rgb_array'))  # update the image with the new frame after taking the action

        plt.axis('off')
        display.display(plt.gcf())  # show the updated frame
        display.clear_output(wait=True)  # clear the previous frame to create an animation effect
```

# CartPole: Keep the tip of the pole straight

```python
env = gym.make('CartPole-v1')  # create the cartpole environment from gym
test_env(env)  # call the function to test the environment
env.close()  # close the environment to free up resources
```

The State

```python
env.observation_space  # gives information about the type and shape of observations in the environment
```

```python
env.action_space  # gives information about the possible actions in the environment
```

# Acrobot: Swing the bar up to a certain height

```python
env = gym.make('Acrobot-v1')  # create the acrobot environment from gym
test_env(env)  # call the function to test the environment
env.close()  # close the environment to free up resources
```

# MountainCar: Reach the goal from the bottom of the valley

```python
env = gym.make('MountainCar-v0') # create the mountaincar environment from gym
test_env(env)
env.close()
```

Pendulum: swing it and keep it upright

```python
env = gym.make('Pendulum-v1') # create the pendulum-v1 environment from gym
test_env(env)
env.close()
```
